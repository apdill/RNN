```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def circle_overlap_area(x1, y1, r1, x2, y2, r2):
    d = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    
    if d >= r1 + r2:
        return 0
    elif d <= abs(r1 - r2) and r1 >= r2:
        return math.pi * r2**2
    elif d <= abs(r1 - r2) and r1 < r2:
        return math.pi * r1**2
    else:
        h = (1/(2*d))*math.sqrt(2*(d**2)*(r1**2)+2*(d**2)*(r2**2) + 2*(r1**2)*(r2**2)-d**4 - r1**4 - r2**4)
        area = (r1**2)*math.asin(h/r1)+(r2**2)*math.asin(h/r2) - d*h
        return area
```
